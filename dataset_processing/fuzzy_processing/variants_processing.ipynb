## Part 1: create pairwise similarity

```python
from fuzzywuzzy import fuzz

import pandas as pd
import json

import numpy as np
import re
import urllib.request 
from itertools import combinations
```

```python
def read_collection(url): # read all_idioms from github
      with urllib.request.urlopen(url) as url:
        data = json.loads(url.read().decode())
        return data

def collect_phrases(data):
    all_phrases = []
    for article in data:
        if len(article['phrase']) != 0:
              all_phrases.append(article['phrase'][0])

    return all_phrases

def modify_phrases(all_phrases):
    modify = lambda string: re.sub(r'[^а-яё]', '', string.lower())
    return [*map(modify, all_phrases)]


def count_ratio_paiwise(all_pairs):
    for p in range(3):
        ratio_list = []
        i = 0
        while i < 78412965:
            pair = next(all_pairs)
            similarity = fuzz.ratio(pair[0], pair[1])
            if similarity >  50:
                ratio_list.append((i, pair, similarity))
            i += 1
            if i % 100000 == 0:
                print(i)
        write_file_with_ratio(ratio_list, p)

def write_file_with_ratio(ratio_list, p):
    with open(f'ratio{p}.txt', 'w') as f:
        f.write(f'{ratio_list}')
```

## Part 2: read ratio files and filter by threshold

```python
def read_ratio(ratio):
    pat = re.compile(r"\d+, \('\w+', '\w+'\), \d+")
    ratio_list = re.findall(pat, ratio)
    for item in ratio_list:
        item = item.split(',')
        return (int(item[0]), (item[1][3:-1],  item[2][2:-2]),  int(item[3][1:]))
```

## Part 3. Join articles belonging to variants

The file variants.txt is now different from the list from the previous section. The reason is that manual filtering of some pairs has been performed: similar phrases, but not semantic variants, were deleted (e.g. *полный рот* and *полный рост*).

```python
with open('variants.txt', 'r', encoding='utf-8') as t:
    variants = t.readlines()
```

```python
# transforming to a convenient format
variants = [*map(lambda x: read_ratio(x), variants)]
```

```python
# number of pairs after manual filtering
len(variants)
```

```python
# list of pairs only (without pair's index and percent of similarity)
all_variants = [item[1] for item in variants]
```

```python
# create a list of indexes corresponding to pairs' items
variants_inds = []

for pair in all_variants:
    pair_ids = []
    for i, phrase in enumerate(modified_phrases):
        if phrase in pair:
            pair_ids.append(i)
    if pair_ids:
        print(pair_ids)
        variants_inds.append(pair_ids)
```

Problem: if there are such pairs like [(x, y), ..., (y, z)], in this case we should combine (x, y, z) into one article, but at this stage we can only get x, y -> (x, y) and y, z -> (y, z) unions.
It is necessary to combine the indexes x, y, z into a single group before combining the articles.

```python
# index joining
to_join_all = sorted(variants_inds, key=lambda x: x[0])
joined_inds = []
i = 0
while i < len(to_join_all[:-1]):
    pair = to_join_all[i]
    pair_2 = to_join_all[i + 1]
    n = 0
    intersection = set(pair).intersection(set(pair_2))
    while intersection:
        n += 1
        pair.extend(set(pair_2).difference(set(pair)))
        pair_2 = to_join_all[i + 1 + n]
        intersection = set(pair).intersection(set(pair_2))
        joined_inds.append(pair)
    else:
        joined_inds.append(pair)
    i += 1 + n
```

```python
joined = list(np.unique(joined_inds))
```
